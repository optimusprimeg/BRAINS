```python
from datetime import datetime
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

Setting the start and end date for dataset

```python
end_date = datetime(year=2024 , month=12, day =31)
start_date = datetime(year=2024 , month=1, day =1)
```

We will download the data for Reliance industrial limited

```python
ticker = "RELIANCE.NS"
data = yf.download(tickers=ticker,
                   start = start_date,
                   end = end_date,
                   interval="1d",
                   auto_adjust= False,
                   progress = False,
                   group_by="ticker"
                   )
data
```

Plotting Candlestickchart for Reliance

```python
import plotly.graph_objects as go
import pandas as pd

fig = go.Figure(data=[go.Candlestick(
    x=data.index,
    open=data[ticker]['Open'],
    high=data[ticker]['High'],
    low=data[ticker]['Low'],
    close=data[ticker]['Close'],
    increasing_line_color='green',
    decreasing_line_color='red'
)])


fig.update_layout(
    title=f'{ticker} Candlestick Chart', #  set the title
    xaxis_title='Date',
    yaxis_title='Price',
    xaxis_rangeslider_visible=False
)


fig.show()
```

```python
data['Daily Return %'] = data[ticker]['Close'].pct_change() * 100

log_returns = np.log(data[ticker]['Close'] / data[ticker]['Close'].shift(1))

data['Log Return %'] = log_returns * 100

data
```

```python
import plotly.graph_objects as go
fig = go.Figure()
# Access 'Daily Return %' directly from the data DataFrame
fig.add_trace(go.Scatter(x=data.index, y=data['Daily Return %'], name='Daily Return %'))
# Access 'Log Return %' directly from the data DataFrame
fig.add_trace(go.Scatter(x=data.index, y=data['Log Return %'], name='Log Return %'))

fig.update_layout(title='Daily Return % vs Log Return %',
                  xaxis_title='Date',
                  yaxis_title='Return',
                  template='plotly_white')
fig.show()
```

```python
import plotly.graph_objects as go

fig_daily = go.Figure()
fig_daily.add_trace(go.Scatter(x=data.index, y=data['Daily Return %'], name='Daily Return %'))

fig_daily.update_layout(title='Daily Return % Over Time',
                        xaxis_title='Date',
                        yaxis_title='Daily Return (%)',
                        template='plotly_white')

fig_log = go.Figure()
fig_log.add_trace(go.Scatter(x=data.index, y=data['Log Return %'], name='Log Return %'))

fig_log.update_layout(title='Log Return % Over Time',
                      xaxis_title='Date',
                      yaxis_title='Log Return (%)',
                      template='plotly_white')

fig_daily.show()
fig_log.show()
```
